4.4 Насколько в среднем стоимость домов с оценкой grade домов выше 10 отличается от стоимости домов с оценкой grade меньше 4?

```python
e = df[df['grade'] > 10]["price"].mean()
h = df[df['grade'] < 4]["price"].mean()
print(e-h)
```
