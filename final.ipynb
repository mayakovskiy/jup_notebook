```python
plt.figure(figsize=(10, 6))
plt.hist(df_h.price)
plt.title("График распределения стоимости недвижимости")
plt.xlabel("Стоимость")
plt.ylabel("Количество")
plt.grid();
```

Вывод: согласно графику можно сделать вывод, об обратной пропорциональности стомости и их количества. Чем больше стоимость недвижимости тем меньше их воличество, ну и соответственно наоборот, что чем меньше стомость - тем большее количество.


```python
plt.figure(figsize=(10,6))
plt.hist(df_h.sqft_living, color="red")
plt.xlabel("квадратные метры жилые")
plt.ylabel("Количетво")
plt.title("График распределения жилых квадратных метров")
plt.grid();
```

Вывод: Преобладающее количество домов с жилой площадью составляет менее 2500 кв.м.

```python
plt.figure(figsize=(10, 6))
plt.hist(df_h.yr_built, color="green")
plt.xlabel("Год строительства дома")
plt.ylabel("Количетво")
plt.title("График распределения по годам строительства домов")
plt.grid();
```

Вывод: дома с годом постройки от 1945 года и позднее - составляют преобладающее количество. Меньшее количество составляют дома которые были построены раньше 1945 года 

Задача 2
1. Изучите распределение домов от наличия вида на набережную
Постройте график
Сделайте выводы
2. Изучите распределение этажей домов
3. Изучите распределение состояния домов

```python
data_w = df_h.waterfront.value_counts()
data_w.index = ['no', 'yes']
data_w
```

```python
plt.figure(figsize=(10,6))
plt.pie(data_w, autopct='%1.1f%%')
plt.title("Распределение домов от наличия набережной")
plt.legend(data_w.index)
plt.show()
```

Вывод: В основной массе домов отсутствует вид на набережную. Меньше чем у 1% домов есть вид на набережную.

```python
stat = df_h.groupby(['floors'], as_index=False)[['id']].count()
stat = stat.rename(columns={"id":"count"})
stat
```

```python
plt.figure(figsize=(10, 6))
plt.bar(stat['floors'], stat['count'], color=['orange', 'red', 'green'], width=0.4)
plt.xlabel("Количество этажей")
plt.ylabel("Количество домов")
plt.title("График распределения этажей домов")
plt.grid();
```

Вывод: Преобладающее количество домов составляют дома с этажностью 1 и 2. Остальные встречаются гораздо реже

```python
df_h.condition.value_counts()
```

```python
stat_c = df_h.groupby(['condition'], as_index=False)[['id']].count()
stat_c = stat_c.rename(columns={"id":"count"})
stat_c
```

```python
plt.figure(figsize=(10, 6))
plt.bar(stat_c['condition'], stat_c['count'])
plt.xlabel('Состояние дома')
plt.ylabel("Количество домов")
plt.title("График распределения состояния домов")
plt.grid()
```

Вывод: Преобладающее количество домов составляет с состоянием дома = 3, т.е. состояние среднее. Далее идут дома с состояними выше среднего. Домов с наихудшим состоянием - минимально.

```python
import seaborn as sns
```

```python
sns.scatterplot(x=df_h["sqft_living"], y=df_h["price"])
plt.title("График зависимости стоимости от жилых квадратных метров")
plt.xlabel("Жилые квадратные метры")
plt.ylabel("Стоимость");
```

На данном графике наблюдается почти прямая зависимость стоимости дома и площади жилых помещений в недвижимости. Чем больше жилых квадратных метров - тем выше стоимость недвижимости.

```python
sns.lineplot(x=df_h["view"], y=df_h["price"])
plt.title("График зависимости стоимости от хорошего вида с окна")
plt.xlabel("Вид из окна")
plt.ylabel("Стоимость");
```

На линейном графике видна зависимость стоимости недвижимости от вида вида с окна. Чем лучше вид с окна - тем выше стоимость недвижимости.

```python
sns.scatterplot(x=df_h["price"], y=df_h["grade"])
plt.title("График зависимости стоимости от конструкции здания и дизайна")
plt.xlabel("Стоимость")
plt.ylabel("Оценка качества конструкции здания и дизайна");
```

На графике видна зависимость стоимости от оценки качества конструкции здания и дизайна. Чем выше качество тем выше стоимость

```python
sat = df_h[df_h["yr_renovated"] != 0]
```

```python
sns.scatterplot(x=df_h["price"], y=sat['yr_renovated'])
plt.title("График зависимости стоимости от года реконструкции здания")
plt.xlabel("Стоимость")
plt.ylabel("Год реконструкции недвижимости");
```

Данный график показывает слабую зависимость стоимости от года реконструкции здания

```python
sns.lineplot(x=df_h["condition"], y=sat['price'])
plt.title("График зависимости стоимости от состояния дома")
plt.xlabel("Состояние дома")
plt.ylabel("Стоимость");
```

На графике видна прослеживается зависимость стоимости дома от его состояния. Чем лучше состояние дома тем выше стоимость
